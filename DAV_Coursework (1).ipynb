```
import pandas as pd
import numpy as np

path = "global_air_quality_data_messy.csv"
df = pd.read_csv(path, parse_dates=['Date'], dayfirst=False, infer_datetime_format=True)

# Basic snapshot
print("Rows:", df.shape[0], "Columns:", df.shape[1])
display(df.head(8))
display(df.info())

# Missing values summary
missing = df.isnull().sum().sort_values(ascending=False)
print("Missing values per column:")
print(missing)

# Unique cities and sample counts
city_counts = df['City'].value_counts().head(15)
print("Top 15 cities by record count:")
print(city_counts)

# Basic numeric statistics
display(df.describe().T)
```

```
import matplotlib.pyplot as plt

# Histogram of PM2.5
plt.figure(figsize=(8,4))
plt.hist(df['PM2.5'].dropna(), bins=50, color='skyblue', edgecolor='black')
plt.title('Distribution of PM2.5')
plt.xlabel('PM2.5 (g/m)')
plt.ylabel('Frequency')
plt.tight_layout()
plt.savefig('fig_pm25_hist.png', dpi=200)  # save screenshot
plt.show()
```

```
# Create a YearMonth column for grouping
df['YearMonth'] = df['Date'].dt.to_period('M')

# Group by month and calculate mean PM2.5
monthly = df.groupby('YearMonth')['PM2.5'].mean().dropna()
monthly.index = monthly.index.to_timestamp()  # convert period to timestamp for plotting

# Plot
plt.figure(figsize=(10,4))
plt.plot(monthly.index, monthly.values, marker='o')
plt.title('Global Monthly Mean PM2.5')
plt.xlabel('Date')
plt.ylabel('PM2.5 (g/m)')
plt.tight_layout()
plt.savefig('fig_monthly_pm25.png', dpi=200)
plt.show()
```

```
# Average PM2.5 per city
city_avg = df.groupby('City')['PM2.5'].mean().dropna()
top10 = city_avg.sort_values(ascending=False).head(10)

# Bar plot
plt.figure(figsize=(10,5))
top10.plot(kind='bar', color='salmon')
plt.title('Top 10 Cities by Average PM2.5')
plt.ylabel('Average PM2.5 (g/m)')
plt.xlabel('City')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('fig_top10_pm25.png', dpi=200)
plt.show()
```

```
import seaborn as sns

# Variables for correlation
vars_for_corr = ['PM2.5','PM10','NO2','SO2','CO','O3','Temperature','Humidity','Wind Speed']
vars_present = [v for v in vars_for_corr if v in df.columns]

# Compute correlation matrix
corr = df[vars_present].corr()

# Heatmap
plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, fmt=".2f", cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation matrix')
plt.tight_layout()
plt.savefig('fig_corr.png', dpi=200)
plt.show()
```

```
# 1. Standardise City
df['City'] = df['City'].astype(str).str.strip()
df['City'] = df['City'].str.replace(r'\s+$', '', regex=True)
df['City'] = df['City'].str.title()  # "london  " -> "London"
```

```
print("Unique cities after cleaning:", df['City'].nunique())
print(df['City'].unique()[:10])
```

```
# clean date
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
before = len(df)
df = df.dropna(subset=['Date'])
after = len(df)

print("Dropped invalid date rows:", before - after)
print("Remaining rows:", len(df))
```

```
# removing duplicates
before = len(df)
df = df.drop_duplicates()
print("Duplicate rows removed:", before - len(df))
```

```
#aggregate duplicated( country+ city code)
agg_cols = ['PM2.5','PM10','NO2','SO2','CO','O3','Temperature','Humidity','Wind Speed']
group_cols = ['City','Country','Date']

before = len(df)
df = df.groupby(group_cols, as_index=False)[agg_cols].mean()
print("Rows merged (duplicate stations collapsed):", before - len(df))
print("New shape:", df.shape)
```

```
# count missing
df['missing_count'] = df[agg_cols].isnull().sum(axis=1)

before = len(df)
df = df[df['missing_count'] < 4].copy()
print("Rows removed where >=4 pollutants were missing:", before - len(df))

# impute by City + YearMonth
df['YearMonth'] = df['Date'].dt.to_period('M')

for col in agg_cols:
    df[col] = df.groupby(['City','YearMonth'])[col].transform(lambda x: x.fillna(x.mean()))

# fallback to full city mean
for col in agg_cols:
    df[col] = df.groupby('City')[col].transform(lambda x: x.fillna(x.mean()))

print("Remaining missing values per column:")
print(df[agg_cols].isnull().sum())
```

```
#Derived features
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month

print(df[['Date','Year','Month']].head())
```

```
#pm2.5 category
def pm25_cat(x):
    if x <= 10: return 'Good'
    if x <= 25: return 'Moderate'
    if x <= 50: return 'Unhealthy for Sensitive Groups'
    if x <= 100: return 'Unhealthy'
    return 'Very Unhealthy'

df['PM25_Category'] = df['PM2.5'].apply(pm25_cat)

print(df['PM25_Category'].value_counts().head())
```

```
#pollution index
from scipy import stats

polls = ['PM2.5','PM10','NO2','SO2','CO','O3']
df_z = df[polls].apply(lambda x: stats.zscore(x.fillna(x.mean())))
df['PollutionIndex'] = df_z.sum(axis=1)

print(df[['City','Date','PollutionIndex']].head())
```

```
df = df.drop(columns=['missing_count'])
df.to_csv("global_air_quality_cleaned.csv", index=False)

print(" Saved cleaned dataset to global_air_quality_cleaned.csv")
print("Final cleaned rows:", len(df))
```

```
import pandas as pd
import matplotlib.pyplot as plt

# Load cleaned dataset
df = pd.read_csv("global_air_quality_cleaned.csv", parse_dates=['Date'])
df['YearMonth'] = pd.to_datetime(df['YearMonth'])

# Monthly average PM2.5
monthly = df.groupby('YearMonth')['PM2.5'].mean()

plt.figure(figsize=(10,4))
plt.plot(monthly.index, monthly.values, marker='o', color='green')
plt.title('Global Monthly Average PM2.5 (Cleaned Data)')
plt.xlabel('Date')
plt.ylabel('PM2.5 (g/m)')
plt.tight_layout()
plt.savefig('cleaned_monthly_pm25.png', dpi=200)
plt.show()
```

```
import plotly.express as px

# Average PM2.5 by country
country_avg = df.groupby('Country')['PM2.5'].mean().reset_index()

fig = px.choropleth(country_avg,
                    locations='Country',
                    locationmode='country names',
                    color='PM2.5',
                    color_continuous_scale='Reds',
                    title='Average PM2.5 by Country',
                    labels={'PM2.5':'Average PM2.5'})
fig.show()
```
